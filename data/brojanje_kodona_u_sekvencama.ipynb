```python
import numpy as np
import pandas as pd
from itertools import product
import os
from Bio import SeqIO
```

```python
nucleotides = ['A', 'C', 'G', 'T']
combinations_of_3 = list(product(nucleotides, repeat=3))

kolone = []
kolone += [''.join(combination) for combination in combinations_of_3]

nepoznate_kolone = []
nepoznate_kolone += ['Y', 'R', 'S', 'W', 'M', 'K', 'H', 'D', 'B', 'V', 'N']

oznake = []
oznake += nucleotides + nepoznate_kolone

print(kolone)
print(nepoznate_kolone)
print(oznake)
```

```python
genoms = ['NC_017627', 'NC_008563']
```

```python
tabele = []
nepoznati_kodoni_svih_genoma = []

for genom in genoms:
    # patogena - 1, nepatogena - 0
    tabela = pd.DataFrame(index=[genom], columns=kolone)
    tabela.fillna(0,inplace=True)
    tabela['Patogenost'] = 0
    for col in tabela.columns:
        tabela[col] = tabela[col].astype(int)
    tabele.append(tabela)

    nepoznati_kodoni_svih_genoma.append([])
    
nepoznati_kodoni_svih_genoma
```

```python
for j, genom in enumerate(genoms):
    tabela = tabele[j]
    nepoznati_kodoni = nepoznati_kodoni_svih_genoma[j]

    fasta_file = f"{genom}/{genom}.fasta"
    sequences = SeqIO.parse(fasta_file, "fasta")

    niska = None
    for record in sequences:
        #print(f"ID: {record.id}")
        #print(record.seq[len(record.seq)-1])
        niska = record.seq
    
    i = 0
    while i != len(niska) - 2:
        kodon = niska[i:i+3]

        # ACT
        if kodon in tabela.columns:
            tabela[kodon] += 1
            
        else:
            print(f'Nepoznata oznaka u kodonu {kodon}')
            ind = 0
            for oznaka in kodon:
                if oznaka not in oznake:
                    print(f'Greska - nepostojeca oznaka {oznaka}')    
                    ind = 1
                    break
            
            if(ind == 0):
                nepoznati_kodoni.append(kodon)
    
        i += 1
    
        if i % 10000 == 0:
            print(i)
    print(f'Tabela broj {j} za genom {genom} uspesno formirana', '\n')

#tabele
```
