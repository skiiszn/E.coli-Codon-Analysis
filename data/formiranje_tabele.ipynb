```python
def mapiraj_tip(row):
    if 'PAI Number' in merged_dataframe.columns and not np.isnan(row['PAI Number']):
        return 0
    elif 'cPAI Number' in merged_dataframe.columns and not np.isnan(row['cPAI Number']):
        return 1
    elif 'nPAI Number' in merged_dataframe.columns and not np.isnan(row['nPAI Number']):
        return 2
    return -1 
```

```python
for i, genom in enumerate(genoms):
    dir_path = f"./{genom}"
    all_dataframes = []
    types = set()
    print(i)
    for filename in os.listdir(dir_path):
        if os.path.isfile(os.path.join(dir_path, filename)):
            if filename.endswith('.csv') and filename != 'patogena_ostrva.csv':
                #print(filename)
                types.add(filename[0])
                
                file_path = os.path.join(dir_path, filename)
                df = pd.read_csv(file_path)
                all_dataframes.append(df)
                
    drop_columns = []
    for type in types:
        drop_columns.append(f"{type if type != 'P' else ''}PAI Number")

    if(len(types) != 0):
        merged_dataframe = pd.concat(all_dataframes, ignore_index=True)
        merged_dataframe.insert(0, 'Type', merged_dataframe.apply(mapiraj_tip, axis=1))
        merged_dataframe = merged_dataframe.drop(drop_columns, axis=1)
        print(merged_dataframe, '\n')
        merged_dataframe.to_csv(os.path.join(dir_path, 'patogena_ostrva.csv'), index=False)
```
