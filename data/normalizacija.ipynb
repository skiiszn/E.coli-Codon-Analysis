```python
def fasta_sequence_lengths(fasta_file):
    with open(fasta_file, 'r') as file:
        sequence_id = None
        sequence = ''
        for line in file:
            line = line.strip()
            if line.startswith('>'):
                if sequence_id:
                    print(f"Sequence {sequence_id} length: {len(sequence)}")
                sequence_id = line[1:]
                sequence = ''
            else:
                sequence += line
        if sequence_id:
            print(f"Sequence {sequence_id} length: {len(sequence)}")
    return len(sequence)

for index, row in konacna_tabela.iterrows():
    if row["Tip"] == "K":
        size = fasta_sequence_lengths(f'{row["ID"]}/{row["ID"]}.fasta')
        konacna_tabela.at[index,"Pocetak"] = 0
        konacna_tabela.at[index,"Kraj"] = size


konacna_tabela
```
