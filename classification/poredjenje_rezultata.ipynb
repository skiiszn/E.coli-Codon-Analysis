```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%store -r knn_best_accuracy
%store -r dtc_best_accuracy
%store -r rfc_best_accuracy
%store -r xgb_best_accuracy
```

```python
models = ['knn', 'dtc', 'rfc', 'xgb']
accuracies = [knn_best_accuracy, dtc_best_accuracy, rfc_best_accuracy, xgb_best_accuracy]
colors = ['orange', 'pink', 'purple', 'brown']
```
